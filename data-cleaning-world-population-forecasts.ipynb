```python
# Create a messagebox template
def msgbox(text):
    """
    Return a massage box template for displaying text within borders
    """
    # Find the maximum line length in the text
    max_line_length = max(len(line) for line in text.split("\n")) + 8

    # Create the top border of the box
    top_border = "╔" + "═" * (max_line_length + 2) + "╗"

    # Create the bottom border of the box
    bottom_border = "╚" + "═" * (max_line_length + 2) + "╝"

    # Create the lines of text inside the box
    text_lines = text.split("\n")
    formatted_lines = ["║ " + line.center(max_line_length) + " ║" for line in text_lines]

    # Combine all the components to create the boxed text
    boxed_text = "\n".join([top_border] + formatted_lines + [bottom_border])
    
    # show the output
    print(boxed_text)
help(msgbox)
```

```python
# Import population data from kaggle
data = pd.read_csv("/kaggle/input/2024-population-projections-by-country/2024Populations.csv")
data
```

```python
# remove unclear columns
data = data.iloc[:,0:-4]
data
```

```python
# Extracting countries and their associated continent from StatisticsTimes.com
url='https://statisticstimes.com/geography/countries-by-continents.php'
country_x_cont = pd.read_html(url)[-1]
country_x_cont = country_x_cont[['Country or Area','Continent']].set_index('Country or Area')
country_x_cont
```

```python
# Insert new column for 'continent'
data.insert(loc=3, column='continent', value='')
data.columns
```

```python
# Mapping countries listed in the dataset to their appropriate continent
data['continent']=''
for i in data['country']:
    for x in country_x_cont.index:
        if i.lower() == x.lower():
            data['continent'].loc[data['country'] == i] = country_x_cont['Continent'].loc[country_x_cont.index == x][0]
        elif i.lower() in x.lower():
            data['continent'].loc[data['country'] == i] = country_x_cont['Continent'].loc[country_x_cont.index == x][0]
        elif i.lower()[:10] == x.lower()[:10]:
            data['continent'].loc[data['country'] == i] = country_x_cont['Continent'].loc[country_x_cont.index == x][0]
        elif i.lower()[-5:] == x.lower()[-5:]:
            data['continent'].loc[data['country'] == i] = country_x_cont['Continent'].loc[country_x_cont.index == x][0]
        else:
            continue
    if i == 'Falkland Islands':
        data['continent'].loc[data['country'] == i] = 'South America'
    elif i == 'Vietnam':
        data['continent'].loc[data['country'] == i] = 'Asia'
    elif i == 'Ivory Coast':
        data['continent'].loc[data['country'] == i] = 'Africa'
    elif i == 'Taiwan':
        data['continent'].loc[data['country'] == i] = 'Asia'
    elif i == 'Laos':
        data['continent'].loc[data['country'] == i] = 'Asia'
    elif i=='Republic of the Congo':
        data['continent'].loc[data['country'] == i] = 'Asia'
    elif i == 'Macau':
        data['continent'].loc[data['country'] == i] = 'Asia'
    elif i == 'Curacao':
        data['continent'].loc[data['country'] == i] = 'North America'
    elif i == 'Vatican City':
        data['continent'].loc[data['country'] == i] = 'Europe'
    else:
        continue
        
# Check to ensure that no 'continent' row is blank            
data[data['continent']==''].reset_index(0)
```

```python
# Extract years from the dataset
years = []
for i in data.columns:
    if 'pop' in i:
        years.extend([i[3:]])
years = pd.DataFrame({'year': years}).astype(int)
years
```

```python
# Extracting the world population data from WorldoMeter.info
url = 'https://www.worldometers.info/world-population/world-population-by-year/'
world_pop = pd.read_html(url)[0][::-1]
world_pop = world_pop[['Year','World Population']]
world_pop
```

```python
# Extracting the world population associated with the years available in the dataset and 
# updating the datset with new columns for the world population
for i in world_pop['Year']:
    for x in years['year']:
        if x==i:
            data[f'wrld_pop{x}'] = int(world_pop["World Population"][world_pop['Year']==i].iloc[0])
        else:
            continue
            
data.sample(10)
```
